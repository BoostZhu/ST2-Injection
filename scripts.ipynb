```python
import os
import shutil
from tkinter import Tk, filedialog

# --- 图形化文件夹选择功能 ---
def select_folder(title_text):
    """
    弹出一个图形化窗口让用户选择文件夹。
    """
    print(f"请在弹出的窗口中选择文件夹：'{title_text}'")
    root = Tk()
    root.withdraw()  # 隐藏主窗口
    root.attributes('-topmost', True)  # 让选择窗口置于顶层
    folder_path = filedialog.askdirectory(title=title_text)
    root.destroy()
    return folder_path

print("库和辅助功能准备就绪。")
```

```python
# --- 主菜单 ---
print("--- 文件/文件夹处理器 ---")
print("您想执行哪个任务？")
print("  1. 删除特定结尾的【子文件夹】")
print("  2. 移动特定结尾的【文件或文件夹】")

task_choice = input("\n请输入任务编号 (1 或 2): ")

# --- 任务 1: 删除子文件夹 ---
if task_choice == '1':
    print("\n--- 开始【删除】任务 ---")
    
    target_folder = input("请输入目标文件夹的【完整路径】: ")
    
    if not os.path.isdir(target_folder):
        print(f"错误：路径 '{target_folder}' 不存在或不是一个文件夹。任务中断。")
    else:
        delete_endings_input = input("请输入要删除的子文件夹【结尾的文字】（多个用空格隔开）: ")
        delete_endings = tuple(delete_endings_input.split())

        print("\n--- 演习模式：检查将要删除的文件夹 ---")
        folders_to_delete = []
        for item_name in os.listdir(target_folder):
            full_path = os.path.join(target_folder, item_name)
            if os.path.isdir(full_path) and item_name.endswith(delete_endings):
                folders_to_delete.append(full_path)
                print(f"  [计划删除] {full_path}")
        
        if not folders_to_delete:
            print("\n没有找到符合条件的文件夹可供删除。")
        else:
            print("\n⚠️ 请仔细检查以上列表！操作不可逆！")
            confirmation = input("确认要删除以上所有文件夹吗？请输入 'yes' 以继续: ")
            if confirmation.lower() == 'yes':
                print("\n--- 开始执行删除 ---")
                for folder_path in folders_to_delete:
                    try:
                        shutil.rmtree(folder_path)
                        print(f"  [成功] 已删除: {folder_path}")
                    except Exception as e:
                        print(f"  [失败] 删除时出错: {e}")
                print("--- 删除任务完成 ---")
            else:
                print("--- 操作已取消 ---")

# --- 任务 2: 移动文件或文件夹 ---
elif task_choice == '2':
    print("\n--- 开始【移动】任务 ---")
    
    source_folder = input("请输入源文件夹的【完整路径】（从哪里移动）: ")
    destination_folder = input("请输入目标文件夹的【完整路径】（移动到哪里）: ")

    if not os.path.isdir(source_folder) or not os.path.isdir(destination_folder):
        print(f"错误：源或目标路径无效。请检查后重试。任务中断。")
    else:
        move_endings_input = input("请输入要移动的项目【结尾的文字】（多个用空格隔开）: ")
        move_endings = tuple(move_endings_input.split())

        print("\n--- 演习模式：检查将要移动的项目 ---")
        items_to_move = []
        for item_name in os.listdir(source_folder):
            if item_name.endswith(move_endings):
                items_to_move.append(item_name)
                print(f"  [计划移动] {item_name}")

        if not items_to_move:
            print("\n没有找到符合条件的项目可供移动。")
        else:
            print(f"\n共 {len(items_to_move)} 个项目将被移动到 '{destination_folder}'")
            print("⚠️ 请仔细检查！")
            confirmation = input("确认要移动以上所有项目吗？请输入 'yes' 以继续: ")
            if confirmation.lower() == 'yes':
                print("\n--- 开始执行移动 ---")
                for item_name in items_to_move:
                    source_path = os.path.join(source_folder, item_name)
                    try:
                        shutil.move(source_path, destination_folder)
                        print(f"  [成功] 已移动: {item_name}")
                    except Exception as e:
                        print(f"  [失败] 移动时出错: {e}")
                print("--- 移动任务完成 ---")
            else:
                print("--- 操作已取消 ---")
else:
    print("\n无效的选择。请重新运行单元格并输入 1 或 2。")
```
